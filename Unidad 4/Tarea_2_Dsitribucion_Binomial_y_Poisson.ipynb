### **Problemas de Tarea Poisson y Binomial**

### **Poisson**

```python
from scipy.stats import poisson
#==============================================================================================================================================================
#¿Cual es la probabilidad de que el numero de personas ingresadas a una unidad de cuidados intensivos en un hospital particular, en un dia particular, sea dos?
#==============================================================================================================================================================

media_personas_ingresadas = 5
X = 2
probabilidad_2_personas_ingresadas = poisson.pmf(X, media_personas_ingresadas)
print("Probabilidad de que dos personas sean ingresadas en un dia es:", probabilidad_2_personas_ingresadas)

probabilidad_mayor_10 = 1 - poisson.cdf(10, media_personas_ingresadas)
print("Probabilidad de que x exceda 10:", probabilidad_mayor_10)
```

### ***Es tipo Poisson***
###Ayuda a modelar datos que representa sucesos de algun evento especificado en una unidad determinada de tiempo o espacio. En esta distribuccion X representa el numero de eventos(numero de personas que ingresan por dia), tambien se cuenta con un promedio por dia, de esta forma se puede saber la probabilidad de que en evento ocurra basado en unidades de tiempo.

### **Binomial**

```python
from scipy .stats import binom
#=======================================================
#¿Cual es la probabilidad de que ninguno sea defectuoso?
#=======================================================

# parametros de la distribuccion binomial
no_satisfactorio = 0.001
num_motores =1000
X = 1000

probabilidad_ningun_defectuoso = binom.pmf(X, num_motores, no_satisfactorio)
print('La probabilidad de que ningun motor sea defectuoso es:', probabilidad_ningun_defectuoso)
```

```python
from scipy .stats import binom
#=======================================================
#¿Cual es la probabilidad de que tres sean defectuosos?
#=======================================================

# parametros de la distribuccion binomial
no_satisfactorio = 0.001
num_motores =1000
X = 3

probabilidad_tres_defectuosos = binom.pmf(X, num_motores, no_satisfactorio)
print('La probabilidad de que tres motores sean defectuosos es:', probabilidad_tres_defectuosos)
```

```python
from scipy .stats import binom
#=======================================================
#¿Cual es la probabilidad de que cuatro sean defectuosos?
#=======================================================

# parametros de la distribuccion binomial
no_satisfactorio = 0.001
num_motores =1000
X = 4

probabilidad_cuatro_defectuosos = binom.pmf(X, num_motores, no_satisfactorio)
print('La probabilidad de que cuatro motores sean defectuosos es:', probabilidad_cuatro_defectuosos)
```

### **Es tipo Binomial**
### Tiene un numero finito de valores, consta de n numeros con una probabilidad p de exito o fracaso para cada numero. En este caso el numero finito es 1000, los n numeros son los valores que tenga X y la probabilidad de exito o fracaso es de 0.001 de que cualquier motor del proveedor resulte no satisfactorio.
