### **Tarea 1 Distribucion Normal**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
plt.figure(figsize=(6,4))

# Parámetros de la distribución normal
media= 70 # Media
desv_std=4 # Desviación estándar
total_estudiantes=500

#Limites de peso
peso_minimo=68
peso_maximo=72

#Variable para cálculo de la probabilidad de que un estudiante que pese menos de 75 kg
peso_limite=72

#Primero necesitamos calcular la rpobaiblidad de que un estudiante pese menos
#de 72 kg utilizando la funcion de distribucion acumulativa (CDF) de la distribucion normal
prob_menos_72kg=norm.cdf(peso_maximo, media, desv_std)

#Probabilidad de pesar menos de 68 kg
prob_menos_68kg=norm.cdf(peso_minimo, media, desv_std)

#Probabilidad de pesar entre 68 y 72 kg
prob_entre_68_72kg=prob_menos_72kg-prob_menos_68kg
```

```python
#Numero de estudiantes que pesan entre 68 y 72 kg
estudiantes_entre_68_72kg=total_estudiantes*prob_entre_68_72kg

print("Probabilidad de que un estudiante pese entre 68 y 72 kg:", prob_entre_68_72kg)
print("Numero de estudianes que pesan entre 68 y 72 kg:", estudiantes_entre_68_72kg)

#Calcular la distribucion normal
x=np.linspace(media- 4*desv_std, media + 4*desv_std, total_estudiantes)
y=norm.pdf(x,media,desv_std)

#Graficar la distribucion normal
plt.plot(x,y,'b', linewidth=2)

#Calculo del numero de estudiantes que pesan menos de 72 kg
estudiantes_menos_72kg=total_estudiantes*prob_menos_72kg
print("Probabilidad de que un estudiante pese menos de 72 kg:", prob_menos_72kg)
print("Número de estudiantes que pesan menos de 72 kg:", estudiantes_menos_72kg)

#Graficar la distribucion normal
x=np.linspace(media-4*desv_std, media+4 * desv_std, 1000)
y=norm.pdf(x, media, desv_std)

#Graficar la distribucion normal
plt.plot(x,y, 'b', linewidth=2)
plt.fill_between( x,y, where=(x >= peso_minimo) & (x <= peso_maximo),color='skyblue', alpha=0.5)
plt.title('Probabilidad de que un estudiante pese entre 68 y 72 kg. \n $\\mu=70$, $\\sigma=4$, x=500')
```
