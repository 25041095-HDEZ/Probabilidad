```python
nombre = ["a", "b", "c", "d", "e", "f", "g"]
n = len(nombre)
r = 3 # ¿cómo agrupar?

Pr = mt.factorial(n) / (mt.factorial(n-r))

print("Existen ", Pr , "posibles permutaciones sin repeticion del total de ",
      n , "nombre ", "en grupos de ", r ," en ", r)

print("Aqui las permutaciones")
permutaciones_nombres = list(permutations(nombre, r))

# Imprime las premutaciones
for permutacion in permutaciones_nombres:
  print(permutacion)
```
