```python
# Basado en el Libro de estadistica para administracion pagina 12
# Creamos una lista vacia que representa la edad de 100 personas
import random
edades = []
for i in range(100):
  numero_aleatorio = random.randint(18, 70)
  edades.append(numero_aleatorio)
  print(edades)
```

```python
import pandas as pd ;
```

```python
# Crea un FataFrame de pandas con los numeros aleatorios
df = pd.DataFrame(edades, columns=['Edad'])
df.head()
```

```python
# Calcular la tabla de distribucion de frecuencias
tabla_frecuencias = df['Edad'].value_counts().reset_index()
tabla_frecuencias.columns = ['Edad', 'Frecuencia']
tabla_frecuencias = tabla_frecuencias.sort_values(by = 'Edad')

print("Tabla de Distribucion de Frecuencias")
tabla_frecuencias
```

Calcular la frecuencia relativa

```python
tabla_frecuencias['Frec Relativa'] = tabla_frecuencias['Frecuencia'] /100
tabla_frecuencias
```

Calcular la frecuencia porcentual

```python
tabla_frecuencias['Frec Porcentual'] = tabla_frecuencias['Frecuencia']/100*100
tabla_frecuencias
```

Calcular la frecuencia acumulada

```python
tabla_frecuencias['Frec Acumulada'] = tabla_frecuencias['Frecuencia'].cumsum()
tabla_frecuencias
```

Reglas de sturges

  1. Es una regla que sirve para calcular el numero de clases o intervalos idóneo en los que se debe dividir un conjunto de datos.
  2. Establece que **el numero de clases es igual o uno más el logaritmi en base dos  del número total de datos.**

***c =1+log2(N)***

donde ***N*** es el tamaño de la muestra

```python
# importamos la librería math de funciones matemáticas
import math
```

```python
# Volvemos a usar la lista edades
print(edades)
```

```python
# Crea un DataFrame con las edades
df = pd.DataFrame({'Edad' : edades})
df
```

```python
# Calcular el número de intervalos utilizando la regla de Sturges
n = len(edades)
K = int( 1 + math.log2(n))
print("Valor de n = ", n, "\nNúmero de intervalos =", K)
```

```python
# Crear la tabla de frecuencias

tabla_frecuencias = pd.cut(df['Edad'], bins=K, include_lowest=True).value_counts().reset_index()
tabla_frecuencias.columns = ['Clase' , 'Frec']
tabla_frecuencias
```

```python
# Ordena la tabla de frecuencia por el intervalo
tabla_frecuencias = tabla_frecuencias.sort_values(by='Clase')
tabla_frecuencias
```

```python
# Generar columnas siguientes
tabla_frecuencias['Frec Rel'] = tabla_frecuencias ['Frec'] / n
tabla_frecuencias['Frec Porc'] = tabla_frecuencias ['Frec'] / n * 100
tabla_frecuencias['Frec Ac'] = tabla_frecuencias['Frec'].cumsum()
tabla_frecuencias['Frec Rel Ac'] = tabla_frecuencias['Frec Rel'].cumsum()
tabla_frecuencias['Frec Porc Ac'] = tabla_frecuencias['Frec Porc'].cumsum()
```
