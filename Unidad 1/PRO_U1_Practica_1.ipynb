#Ejemplo 1

```python
# Lista de calificaciones del grupo A de probabilidad y estadistica
GrupoA = [70, 70, 65, 90, 100, 95, 45, 80, 73, 20, 100, 96]
print(GrupoA)
type(GrupoA)
```

```python
# Sacar la media de las calificaciones con una formula
promedio = sum(GrupoA) / len(GrupoA)
print(promedio)
```

```python
#Utilizar la funcion mean que esta en la libreria statistics para sacar
# el promedio
promedio2 =mean(GrupoA)
print(promedio2)
```

```python
# Determina el valor de la media con la funcion median

mediana = median(GrupoA)
print(mediana)
```

```python
# Determina el valor de la moda con la funcion mode

moda = mode(GrupoA)
print(moda)
```

```python
#En caso de que exista mas de una moda, se puede completar la funcion multimode para sacar
# Multiples modas que pueden existir
moda2 = multimode(GrupoA)
print(moda2)
```

```python
import random
import matplotlib.pyplot as plt
from statistics import *
```

```python
fabrica1 = []
for i in range(100):
  numero_aleatorio = round( random.uniform(50, 100), 2)
  fabrica1.append(numero_aleatorio)
print(fabrica1)
```

```python
fabrica2 = []
for i in range(100):
  numero_aleatorio = round( random.uniform(50, 100), 2)
  fabrica2.append(numero_aleatorio)
print(fabrica2)
```

## Maximo, minimo y rango

```python
maximo1 = max(fabrica1)
minimo1 = min(fabrica1)
rango1 = maximo1 - minimo1
media1 = mean(fabrica1)
mediana1 = median(fabrica1)
moda1 = mode(fabrica1)

print("Estadistica de la fabrica1 \n", "Media =", media1)
```

```python
desviacion_estandar_fabrica1=stdev(fabrica1)
desviacion_estandar_fabrica2=stdev(fabrica2)
```

```python

plt.figure(figsize = (8,6) )
plt.hist(fabrica1, bins=10, alpha=0.2, color="b", edgecolor="black")
plt.axvline(media1, color='b', linestyle='dashed', linewidth=2, label=f'Media={media1:.2f}')
plt.axvline(mediana1, color='r', linestyle='dashed', linewidth=2, label=f'Mediana={mediana1:.2f}')
plt.axvline(moda1, color='y', linestyle='dashed', linewidth=2, label=f'Moda={moda1:.2f}')
plt.axvline(media1 - desviacion_estandar_fabrica1, color='black', linestyle='dashed',
            linewidth=2, label=f'Desv. Estandar ={desviacion_estandar_fabrica1:.2f}')
plt.axvline(media1 - desviacion_estandar_fabrica1, color='black', linestyle='dashed',
            linewidth=2)

plt.xlabel('Valor')
plt.ylabel('Frecuencia')
plt.title("Produccion diaria de la fabrica1")
plt.legend()
plt.show()
```
