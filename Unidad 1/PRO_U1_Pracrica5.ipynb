En esta practica el alumno usará **conceptos de la estadisticadescriptiva,** las medidas de tendencia central y la **visualizacion de datos** y los aplicara en un conjunto de datos real



*   El archivo a utilizar es una base de datos publica por la secretaria de salud sobre las pruebas realizadas por el COVID
*   El archivo es de texto tipo CSV (Comma Separated Values)



```python
import pandas as pd
import matplotlib as plt
import seaborn as sns
from matplotlib import pyplot
```

```python
# Carga el archivo CSV en una variable llamada covid
covid = pd.read_csv("Covid.csv", encoding = "ISO-8859-1")
```

```python
# Mostramos los primeros 5 registros del archivo
covid.head()
```

```python
# Mostramos la informacion del archivo
# Muestra el numero de registros, tipos de datos de la columna, valores nulos, memoria utilizada, etc
covid.info()
```

```python
# Mostrar la edad maxima de la base de la columna edad
covid.EDAD.max()
```

```python
# Mostrar la edad minima de la base de la columna edad
covid.EDAD.min()
```

```python
# Dibuja el histograma del atributo "EDAD" con bins = 3
# Calcular la media y guardarla en una variable para mostrarla en el histograma
media = covid.EDAD.mean()
plt.pyplot.hist(covid.EDAD, bins = 11)
plt.pyplot.title("Histograma Edad Pruebas COVID")
plt.pyplot.axvline(media, color='y', linestyle='dashed', linewidth=2, label=f'Media={media:.2f}')
plt.pyplot.xlabel("Edad")
plt.pyplot.ylabel("Frecuencia")
```

```python
# Dibuja el histograma de la columna sexo,
# esto premite comparar cuantas hay de cada grupo
plt.pyplot.hist(covid.SEXO, bins = 3)

# establece las etiquetas x/y y muestra el titulo
plt.pyplot.title("Histograma Sexo de Pruebas COVID")
plt.pyplot.xlabel("Sexo")
plt.pyplot.ylabel("Frecuencia")
```

```python
# Dibuja el histograma del atributo "Tipo paciente, esto permite
# comparar cuantos hay de cada grupo
plt.pyplot.hist(covid.TIPO_PACIENTE, bins = 3)

# establece las etiquetas x/y y muestra el título
plt.pyplot.title("Histograma de tipos Pacientes en Pruebas COVID")
plt.pyplot.xlabel("Tipo de Paciente")
plt.pyplot.ylabel("Frecuencia")
```

```python
# Crear un diagrama de cajs para visualizar la distribucion
# de los rangos de edad por sexo
sns.boxplot(x="SEXO", y ="EDAD", data=covid)
```

```python
# Crear un diagrama de cajs para visualizar la distribucion de
# los rangos de edad por tipo de paciente
sns.boxplot(x="TIPO_PACIENTE", y ="EDAD", data=covid)
```

```python
# Crear un diagrama de cajas para visualizar la ditribucion de
# los rangos de edad por tipo de participante

sns.boxplot(x="RESULTADOS", y ="EDAD", data=covid)
```
